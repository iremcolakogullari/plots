# Histogram

```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

```python
sns.histplot(df.price, kde = False, color = 'purple');
```

```python
sns.histplot(df.price, bins = 10, kde = False, color = "purple");
```

```python
sns.distplot(df.price, kde = True, hist = False);
```

```python
sns.kdeplot(df.price, shade = False);
```
